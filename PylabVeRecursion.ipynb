## Bir fonksiyon grafiği çizme

```python
x = linspace(0,10,100) # 0 ile 10 arasında eşit aralıklı 100 tane eleman oluşturur
```

```python
y = sin(x) # x değerlerini kullanarak 100 elemanlı y dizisi oluşturur.
```

```python
plot(x,y)
xlabel('x değerleri')
ylabel('y =sin(x)')
```

### Alıştırma: Sin(x) ve Cos(x) fonksiyonlarını aynı pencerede çizdirin. $x$'in değeri [0,10] arasında değişsin ve bu aralıkta eşit aralıklı 1000 değer oluştursun.

```python
x = linspace(0,10,1000)
y1 = sin(x)
y2 = cos(x)
```

```python
plot(x,y1,'g--', label='sine') # farklı renleri dene. g = green, y = yellow.
plot(x,y2,'y-.',label='cosine') # Çizgi biçimini değiştir! "--", ":", "-"
xlabel(" x ekseni")
ylabel(" y = sin x or y = cos x")
legend(loc='upper right')
```

###  Ödev: $ y = (x-3)^2 -12$ fonksiyonunu Python'da yazın.   Bu fonksiyonu [-10,10] arasında çizdirin ve bu aralıkta 100 tane değer olsun.  Grafik  kırmızı renkte ve kesikli renkli olsun. 

```python
def f(x):
    return (x-3)**2 - 12
```

```python
x = linspace(-10,10,100)
y = f(x)
plot(x,y, 'r--')
grid(True)
```

###  Aşağıdaki $x$ ve $y$ değerlerini kullanarak *deltoid" eğrisini çizin. 
$$ x = 2\cos(\theta) + cos(2\theta)   \quad y = 2 \sin(\theta) -sin(2\theta)$$
$\theta$ değeri $ 0\leq \theta \leq 2\pi$ arasında değişsin. 

```python
theta = linspace(0,2*pi,100)
x = 2*cos(theta)+ cos(2*theta)
y = 2*sin(theta)-sin(2*theta)
plot(x,y)
```
